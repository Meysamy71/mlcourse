### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Step 2. Import the dataset 

```python
pd.read_csv('titanic.csv')
```

### Step 3. Assign it to a variable titanic 

```python
titanic = pd.read_csv('titanic.csv')
titanic
```

### Step 4. Set PassengerId as the index 

```python
titanic = titanic.set_index('PassengerId')
titanic
```

### Step 5. Create a pie chart presenting the male/female proportion

```python
GenderChart = titanic.Sex.value_counts(normalize=True) * 100
Mylabel = ['Male', 'Female']
plt.pie(GenderChart, labels=Mylabel, autopct='%1.1f%%', colors=['DodgerBlue', 'DeepPink'])
plt.title('Proportion (%) of male and female')
plt.legend(title='Gender')
plt.show()
```

### Step 6. Create a scatterplot with the Fare payed and the Age, differ the plot color by gender

```python
FarePayed = titanic
FarePayed.Age.fillna(100, inplace=True)

plt.scatter(FarePayed[FarePayed['Sex'] == 'male'].Age,FarePayed[FarePayed['Sex'] == 'male'].Fare,label='Male')
plt.scatter(FarePayed[FarePayed['Sex'] == 'female'].Age,FarePayed[FarePayed['Sex'] == 'female'].Fare,label='Female')

plt.title("Fare payed and the Age")
plt.xlabel("Age in years")
plt.ylabel("Passenger fare")
plt.legend()
plt.show()
```

### Step 7. How many people survived?

```python
titanic.Survived.value_counts().iloc[1]
```

### Step 8. Create a histogram with the Fare payed

```python
sns.set(style="white")
colors = sns.color_palette("viridis", n_colors=1)

plt.figure(figsize=(10, 6))
sns.histplot(titanic['Fare'], kde=True, color=colors[0])

sns.despine(left=True, bottom=True)

plt.title("Fare Histogram", fontsize=18, color='darkgreen')
plt.xlabel("Fare", fontsize=14, color='darkgreen')
plt.ylabel("Density", fontsize=14, color='darkgreen')

plt.xticks(color='orange')
plt.yticks(color='orange')

plt.show()
```
